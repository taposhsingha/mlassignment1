```python
train_data = '/train'
test_data='/test'
val_data = '/val'

```

```python
import matplotlib.pyplot as plt
import random
import os
fig = plt.figure()
random_image = random.choice(os.listdir('/Users/Taposh/desktop/sample_project_1'+'/image_data'+train_data+'/NORMAL'))
img = plt.imread('/Users/Taposh/desktop/sample_project_1'+'/image_data'+train_data+'/NORMAL/'+random_image)
ax = fig.add_subplot(1,1,1)
ax.imshow(img)
ax.set_title('showing a random image from train set')
plt.show
```

```python

train_data_generator = ImageDataGenerator(
rescale=1./255,
rotation_range=10)
validation_data_generator = ImageDataGenerator(rescale=1./255)
test_data_generator = ImageDataGenerator(rescale=1./255)
```

```python
train_data_generate_final =  train_data_generator.flow_from_directory(
'/Users/Taposh/desktop/sample_project_1'+'/image_data'+train_data,
target_size = (200,200),
batch_size = 32,
color_mode = 'grayscale',
class_mode = 'binary',
)

test_data_generate_final =  test_data_generator.flow_from_directory(
'/Users/Taposh/desktop/sample_project_1'+'/image_data'+test_data,
target_size = (200,200),
batch_size = 32,
color_mode = 'grayscale',
class_mode = 'binary',
)

validation_data_generate_final =  validation_data_generator.flow_from_directory(
'/Users/Taposh/desktop/sample_project_1'+'/image_data'+val_data,
target_size = (200,200),
batch_size = 32,
color_mode = 'grayscale',
class_mode = 'binary',
)
```

```python
batch_size = 16
nb_classes =4
nb_epochs = 5
img_rows, img_columns = 200, 200
img_channel = 3
nb_filters = 32
nb_pool = 2
nb_conv = 3
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), padding='same', activation=tf.nn.relu,
                           input_shape=(200, 200, 1)),
    tf.keras.layers.MaxPooling2D((2, 2), strides=2),
    tf.keras.layers.Conv2D(32, (3,3), padding='same', activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D((2, 2), strides=2),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation=tf.nn.relu),
    tf.keras.layers.Dense(4,  activation=tf.nn.softmax)
])
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train_data_generate_final,epochs=10,validation_data=validation_data_generate_final)
```

```python
model.evaluate(test_data_generate_final)
```
